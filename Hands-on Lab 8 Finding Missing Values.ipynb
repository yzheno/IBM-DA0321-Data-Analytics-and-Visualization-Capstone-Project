<h2>1. Load the Dataset</h2>

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/UDKAZw-kz18Yj8P6icf_qw/survey-data-duplicates.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
print(df.info())
df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
missing_data = df.isnull()
missing_data
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
plt.figure(figsize=(10, 8))
sns.heatmap(missing_data, cbar=True, cmap='coolwarm')
plt.title('Missing Values Heatmap')
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
missing_data['Employment'].value_counts()
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
df['Employment'].value_counts().idxmax()
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
df['Employment'] = df['Employment'].replace(np.nan, df['Employment'].value_counts().idxmax())
df
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
# add Count as new column into duplicates dataframe after using value_counts()
counts_df = df['Employment'].value_counts().reset_index(name='Count')
## remove count < = 100
emp = counts_df[counts_df["Count"] > 100]["Employment"]
## filter df employment category to find the same employment categories in emp dataframe
filteredemp = df[df['Employment'].isin(emp)]['Employment']
filtered = filteredemp.value_counts().reset_index()
filtered.columns = ["Employment", "Count"]
plt.figure(figsize=(20, 10))
sns.barplot(x='Count', y='Employment', data=filtered, orient='h')
plt.title('Distribution of Employment')
plt.xticks(rotation=90)
plt.show()
```
